Два бегуна стартовали один за другим с интервалом в 2 мин. Второй бегун догнал первого на расстоянии 1 км.
от места старта, а пробежав от места старта 5 км., он повернул обратно и встретился с первым бегуном через 20
мин. после старта первого бегуна. Найти скорость второго бегуна.

```python
from sympy.abc import x, y
from sympy import Eq
from sympy import solve

sol = solve([Eq(1 / x - 1 / y, 1 / 30), Eq(10 - 1 / 3 * x, 3 * y / 10)], [x, y])
print("v1=", sol[0][0], " km / h", sep="")
print("v2=", sol[0][1], " km / h", sep="")
```
